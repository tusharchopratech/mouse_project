```python
import pandas as pd
import io
import numpy as np
from scipy.signal import butter, lfilter
import matplotlib.pyplot as plt
import os
```

```python
def butter_bandpass(lowcut, highcut, fs, order=4):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data,  lowcut=110, highcut=250, fs=2048, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y
```
